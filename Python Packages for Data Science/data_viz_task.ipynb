```python
# Import libraries
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

%matplotlib inline
```

```python
# Load data
car_data = pd.read_csv('Cars93.csv', index_col = 0, keep_default_na=False) 
```

### 1. Boxplot

```python
# Boxplot for revs per mile
# list of manufacturers
list = ['Audi', 'Hyundai', 'Suzuki', 'Toyota']

# empty array
arr = []
# loop through the list of manufacturers
for brand in list:
    # get values for revs.per.mile in a list
    data = car_data[car_data['Manufacturer'] == brand]['Rev.per.mile'].values
    # add list to empty array to create 2d data.
    arr.append(data)

# boxplot for revs per mile of manufacturers in list.
plt.boxplot(arr)
plt.xlabel('Car Manufacturers')
plt.ylabel('Revs per mile')
plt.title('Revs per mile boxplot for four car manufacturers')
plt.xticks([1, 2, 3, 4], list)
plt.show()
plt.close()

```

**Boxplot question:**  
Which of these manufacturers has the car with the highest revs per mile?


**Boxplot answer:** 
Toyota.

### 3. Lineplot

```python
# Lineplot (Wheelbase vs turning circle)
plt.figure()
fg = sns.lineplot(x = car_data['Wheelbase'], y = car_data['Turn.circle'])
plt.ylabel('Turning circle')
plt.title('Wheelbase vs Turning Circle Lineplot')
plt.show()
```

**Lineplot question:**  
What is this relationship? What happens when the wheelbase gets larger?


**Lineplot answer:**  
Turning circle increases (on average) with increase in wheelbase. 

### 4. Barplot

```python
# Barplot for mean horsepower
plt.figure()
fg = sns.barplot(x = car_data['Type'], y = car_data['Horsepower'])
plt.title('Mean horsepower for different types of cars')
plt.show()
```

**Barplot question:**  
Does a larger car mean more horsepower?



**Barplot answer:** No. Midsize and large cars have similar mean horsepower. Small cars have the minimum average horsepower, but this trend does not continue over all car sizes.
