```python
# Import pandas
import pandas as pd
```

```python
# Create a DataFrame with balance.txt
df = pd.read_csv('balance.txt', sep=' ')
```

Write the code needed to produce a report that provides the following information:
* Compare the average income based on ethnicity. 
* On average, do married or single people have a higher balance?
* What is the highest income in our dataset?
* What is the lowest income in our dataset?
* How many cards do we have recorded in our dataset? (Hint: use `sum()`)
* How many females do we have information for vs how many males? (Hint: use `count()`For a list of all methods for computation of descriptive stats, explore the [pandas documentation](https://pandas.pydata.org/pandas-docs/stable/reference/frame.html#computations-descriptive-stats)). 




```python
# compare average income based on ethnicity
ethnicity_grouped = df.groupby(["Ethnicity"])
print(ethnicity_grouped['Income'].mean())
print("African-American people have a higher income.")
```

```python
# On average, do married or single people have higher balance
marr_or_sing = df.groupby(['Married'])['Balance'].mean()
print(marr_or_sing)
print('Single people have slightly higher average balance.')
```

```python
# What is the highest income in our dataset?
highest_income = df['Income'].max()
print(f"The highest income in our dataset is {highest_income}")
```

```python
# What is the lowest income in out dataset?
lowest_income = df['Income'].min()
print(f'The lowest income in our dataset is {lowest_income}')
```

```python
# How many cards do we have recorded in our dataset? 
total_cards = df['Cards'].sum()
print(f"We have a total of {total_cards} recorded in our dataset.")
```

```python
# How many females do we have information of and how many males?
group_by_gender = df.groupby(['Gender']).count()
print(group_by_gender)
# The database has complete information about 207 females and 193 males with no null values.
```
